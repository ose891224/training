```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 1. 데이터 준비
data_set = np.loadtxt("ThoraricSurgery3.csv", delimiter=",")
print(type(data_set))
print(data_set.shape)
print(data_set.dtype)
```

```python
X = data_set[:, :-1]
X # train data
```

```python
y = data_set[:, 16]
y # label, target
```

```python
print(X.shape)
print(y.shape)
```

```python
#2.모델의 표현방법(NN)
model = Sequential()
model.add(Dense(30, input_dim=16, activation ='relu', name='layer1' ))  # input_dim은 은닉층에만 있

dense = Dense(1, activation='sigmoid', name='layer2')
model.add(dense)
#model.summary()

dense.get_weights()
```

```python
# 모델의 환경 설정
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# 기계학습
hist = model.fit(X, y, epochs=100, batch_size=10)
```
