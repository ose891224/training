```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [[2, 0], [4, 0], [6, 0], [8, 1], [10, 1], [12, 1], [14, 1]]

x_data = [i[0] for i in data]
y_data = [i[1] for i in data]
print(x_data, y_data)
```

```python
X = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(X, y)
plt.show()
```

```python
a = 0 # 기울기: S곡선의 기울기 담당
b = 0 # y의 절편: S곡선의 좌우 이동

lr = 0.05 # 학습률
epochs = 2001

#################################################
# 시그모이드 함수: y축의 범위 0 ~ 1
def sigmoid(x):
  return 1 / (1 + np.e **(-x))

#################################################
# 경사 하강법
for i in range(epochs):
  for x, y in data:
    # 예측값: x * sigmoid( a*x + b ): x의 예측값
    # 오차: 예측값 - 실제값
    # x * sigmoid(a * x + b) - y

    # a의 오차
    a_diff = x * (sigmoid(a * x + b) - y)
    # b의 오차
    b_diff = sigmoid(a * x + b) - y

    # a와 b의 값 업데이트
    a = a - lr * a_diff
    b = b - lr * b_diff

    if i % 100 == 0:
      print("ephochs= %4d, a: %.2f, b: %.2f"%(i, a, b))
```

```python
print(x_data, y_data)
print("a: ", a)
print("b: ", b)
```

```python
print("실제값: ", y_data)
# y_data의 예측값
y_pred = [sigmoid(a * row[0] + b) for row in data] # x_data값
print("예측값: ", y_pred)
```

```python
plt.scatter(x_data, y_data)
plt.scatter(x_data, y_pred)
plt.plot(x_data, y_pred, "red")
plt.show()
```
