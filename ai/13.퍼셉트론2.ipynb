```python
import numpy as np
```

```python
X_data = np.array([[0,0],[0,1],[1,0],[1,1]])
X_data
```

```python
def AND(x):
  w = np.array([0.5,0.5])
  b = -0.7

  ws = (x * w).sum() + b
  print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
for x in X_data:
  print("AND(%d,%d)=%d" %(x[0],x[1],AND(x)))
```

```python
def OR(x):
  w = np.array([0.5,0.5])
  b = -0.2

  ws = (x * w).sum() + b
  print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
for x in X_data:
  print("OR(%d,%d)=%d" %(x[0],x[1],OR(x)))
```

```python
def NAND(x):
  w = np.array([-0.5,-0.5])
  b = 0.7

  ws = (x * w).sum() + b
  print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
for x in X_data:
  print("NAND(%d,%d)=%d" %(x[0],x[1],NAND(x)))
```
