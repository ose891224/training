```python
# 2. **인수: dict타입, 키워드인수
def func(width, height, **kw):
  print(width," ", height)
  print(kw, type(kw))
```

```python
def func(width, height, **kw):
  print(width, " ", height)
  print(kw, type(kw))
  print(kw.keys())
  print(kw.values())
```

```python
func(10, 20)

func(height=200, width=100, depth=5, demension=3)
```

First Class Function: (일등급 함수) 함수 이름을 변수처럼 사용 가능
1. 함수 객체를 다른 함수의 인수로 전달 가능
2. 함수 객체를 다른 함수에서 반환 값으로 전달 가능
3. 함수 객체를 다른 자료구조에 저장 가능

```python
def add_two(a, b):
  print("add_two is called!!")
  return a + b
```

```python
print(type(add_two))
```

```python
b = add_two
print(type(b))
b(10, 20)
```

```python
def func_two(func, a, b):
  print("func_two is called!!")
  result = func(a, b)
```

```python
func_two(add_two, 10, 20)
```

```python
# First Class Function 2
def foo2():
  print("foo 2....")
  # local 함수
  def bar2():
    print("bar 2...")
  return bar2
```

```python

# 함수 내의 로컬 함수는 외부에서 호출할 수 없다
# bar2()

ret = foo2()
print(ret)
```
