**최소제곱법**<br>
기울기 a를 구하는 수식<br>
(X - X평균) * (y - y평균)의 합 / (X - X평균)의합<br>

y평균 - (x의 평균 * 기울기)

```python
# a = (X - X평균) * (y - y평균)의 합 / (X - X평균의합)

X - X.mean()
```

```python
((X - X.mean()) **2).sum() # 분모 (X - X평균의합)
```

```python
# a = (X - X평균) * (y - y평균)의 합 / (X - X평균의합)

a = ((X -X.mean())*(y - y.mean())).sum() / ((X - X.mean()) **2).sum()
a
```

```python
# b = y평균 - (x의 평균 * 기울기)
b = y.mean() - (X.mean() * a)
print(b)
```

```python
y_pred = a * X + b
print(y_pred)
```

```python

# 판 크기 줄이기
plt.figure(figsize=(4, 3))
# 점차트 그리기
plt.scatter(X, y)
# 선챁트 그리기
plt.plot(X, y_pred, "r")
plt.show()
```
