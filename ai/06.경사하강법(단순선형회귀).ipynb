```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
data
```

```python
np.array(data)
```

```python
X_data = [row[0] for row in data]
Y_data = [row[1] for row in data]
print(X_data)
print(Y_data)
```

```python
X = np.array(X_data)
y = np.array(Y_data)
print(X, y)
```

```python
# 판 크기 줄이기
plt.figure(figsize=(4, 3))
# 점차트 그리기
plt.scatter(X, y)
# 선챁트 그리기
# plt.plot(X, y)
plt.show()
```

```python
# 경사하강법
a = b = 0
print(a, b)

lr = 0.03 # 학습률: 이동보폭
epochs = 2001 # 반복횟수

n = len(X)

# 경사 하강법
for i in range(epochs):
  # 예측값: y_pred = ax + b
  y_pred = a * X + b
  # 오차: 실제값 - 예측값
  error = y - y_pred

  # a를 편미분: 2/n * Σ (-Xi * (yi - (aXi + b)))
  a_diff = (2/n) * sum(-X * error)

  # b를 편미분
  b_diff = (2/n) * sum(-(error))

  a = a - lr * a_diff # 학습율에 편미분 값을곱해서 a에서 뺌

  b = b - lr * b_diff
  if i % 100 == 0:
    # print("epochs: ", i,"\t예측값: ", y_pred,"\t기울기: ",a, "\t절편: ", b)
    print("epochs=%.f, 기울기=%.04f, 절편=%.04f"%(i, a, b))
```

```python
# 예측값
y_pred = a * X + b
print(y_pred)

plt.figure(figsize=(4, 3))
plt.scatter(X, y)
plt.plot(X, y_pred, "red")
plt.scatter(X, y_pred)
plt.show()
```
