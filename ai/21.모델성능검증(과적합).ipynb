```python
import pandas as pd
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("sonar.csv", header=None)
# df
```

```python
df.info()
```

```python
df.value_counts()
```

```python
df[60].value_counts()
```

```python
dataset = df.values
dataset
```

```python
dataset.dtype
```

```python
X = dataset[:,:-1]
print(X.shape," : ", X.dtype," : ", X.ndim)

y_obj = dataset[:, 60]
print(y_obj.shape," : ", y_obj.dtype, " : ", y_obj.ndim)
```

```python
X = X.astype(np.float64)

print(X.shape," : ", X.dtype," : ", X.ndim)
```

```python
# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(y_obj)

y = e.transform(y_obj)
y
```

```python
y.dtype
```

```python
# 2. 모델 설정(NN)
nn = Sequential()

nn.add(Dense(24, input_dim=60, activation='relu'))
nn.add(Dense(10, activation='relu'))
nn.add(Dense(1, activation='sigmoid'))
nn.summary()
```

```python
# 모델 컴파일: 딥러닝을 구동하기 위한 환경설정
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# 모델 실행(모델을 기계학습)
nn.fit(X, y, epochs=200, batch_size=5)
```

```python
nn.evaluate(X, y)
```

```python
# 학습 set과  테스트set의 분리
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, t_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape)
print(y_train.shape, t_test.shape)
```

```python
# 인공신경망 모델
model = Sequential()
model.add(Dense(32, input_dim=60, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=200, batch_size=20)
```
