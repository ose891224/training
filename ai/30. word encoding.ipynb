```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
sentences = ['I love my dog.', 'I love my cat.', 'You love my dog!',
             'Do you think my dog is amazing?']

tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')
tokenizer.fit_on_texts(sentences)
tokenizer.word_counts
```

```python
word = tokenizer.word_index
word
```

```python
tokenizer.texts_to_sequences(sentences)
```

```python
test = ['I really love my dog', 'my dog loves my lizard']
tokenizer.texts_to_sequences(test)
```

```python
sequneces = tokenizer.texts_to_sequences(sentences)
sequneces
```

```python
padding = pad_sequences(sequneces)
padding
```

```python
sent = ['I really think this is amazing. honest.']
seq = tokenizer.texts_to_sequences(sent)
print(seq)
onehot = to_categorical(seq) # 다중분류
onehot
```
