```python
import tensorflow as tf
```

```python
scalar = tf.constant([1,2,3])
print(scalar)
print(scalar.shape, scalar.dtype)
print(tf.rank())
```

```python
metrix = tf.constant([[1,2,3], [4,5,6], [7,8,9]])
print(metrix)
print(tf.rank(metrix))
```

```python
tensor = tf.constant([[[1,2,3], [4,5,6], [7,8,9]],
                      [[1,2,3], [4,5,6], [7,8,9]],
                      [[1,2,3], [4,5,6], [7,8,9]]])
print(tensor) # shap = (면, 행, 열)
print(tf.rank(tensor))
```

```python
print(metrix.shape, metrix.dtype)
print(tensor.shape, tensor.dtype)
```

```python
from tensorflow import keras
```

```python
X = np.array([-1.0, 0.0, 1.0, 2.0,3.0,4.0], dtype=np.float32)
X
```

```python
y = np.array([-3.0, -1.0, 1.0, 3.0,4.0,7.0], dtype=np.float32)
y
```

```python
# 모델 생성
model = keras.Sequential([keras.layers.Dense(1, input_shape=[1])])
model
```

```python
model.compile(loss='mean_squared_error', optimizer='sgd')
model.fit(X, y, epochs=500)
```
