```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
x_data
```

```python
def AND(x):
  w = np.array([0.5,0.5])
  b = -0.7

  ws = (x * w).sum() + b
  # print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
def OR(x):
  w = np.array([0.5,0.5])
  b = -0.2

  ws = (x * w).sum() + b
  # print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
for x in x_data:
  print("AND(%d, %d) = %d"%(x[0], x[1], AND(x)))
```

```python
for x in x_data:
  print("OR(%d, %d) = %d"%(x[0], x[1], OR(x)))
```

```python
def NAND(x):
  w = np.array([-0.5,-0.5])
  b = 0.7

  ws = (x * w).sum() + b
  # print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
for x in x_data:
  print("NAND(%d, %d) = %d"%(x[0], x[1], NAND(x)))
```

위의 예제는 각 함수를 통해 다층 퍼셉트론처럼 만들었지만 다층 퍼셉트론이 아님.<br>
각 함수의 w, b값을 지우면 함수의 내용이 동일<br>
다층 퍼셉트론에서 공통으로 사용하는 함수->활성화 함수

```python
def GATE(x, w, b):
  ws = (x * w).sum() + b
  # print(ws)

  if ws >= 0:
    return 1
  return 0
```

```python
for x in x_data:
  print("GATE(%d, %d) = %d"%(x[0], x[1], GATE(x, np.array([0.5, 0.5]), -0.7)))
```

```python
w1 = np.array([[-0.5, -0.5], # NAND weight
               [0.5, 0.5] # OR weight
               ])
b1 = np.array([0.7, # NAND bias
               -0.2 # OR bias
               ])

w2 = np.array([0.5, 0.5]) # AND weight
b3 = np.array([-0.7]) # AND bias
```

```python
print("x1\tx2\tn1\tn2\tyout")
print("-----------------------------------------")
for x in x_data:
  n1 = GATE(x, w1[0], b1[0]) # NAND
  n2 = GATE(x, w1[1], b1[1]) # OR
  n = np.array([n1, n2]) # 은닉층의 입력
  y = GATE(n, w2, b3) # AND
  print("%d\t%d\t%d\t%d\t%d"%(x[0], x[1], n1, n2, y))

```
