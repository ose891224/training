```python
import pandas as pd
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("sonar.csv", header=None)

dataset = df.values
dataset

dataset.dtype

X = dataset[:,:-1]
print(X.shape," : ", X.dtype," : ", X.ndim)

y_obj = dataset[:, 60]
print(y_obj.shape," : ", y_obj.dtype, " : ", y_obj.ndim)

X = X.astype(np.float64)

print(X.shape," : ", X.dtype," : ", X.ndim)

# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(y_obj)

y = e.transform(y_obj)
y
```

```python
# 학습 set과  테스트set의 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, t_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape)
print(y_train.shape, t_test.shape)
```

```python
# 인공신경망 모델
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
# model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# model.summary()

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=200, batch_size=20)
```

```python
result = model.evaluate(X_test, t_test)
print("오차:", result[0])
print("정확도:", result[1])

# 학습데이터의 정확도는 100%에 가깝고 테스트셋의 정확도는 떨어짐
```

```python
model.save("my_model.keras")
```
