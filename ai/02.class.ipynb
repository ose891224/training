1. 클래스(틀): class로 정의, 멤버, 메서드를 가지는 객체

6. 속성: 멤버, 메서드 정의를 가르킴
7. 상속: 상위(부모)클래스의 자원을 자식 클래스가 공유하여 사용

## OOP특징
Object Oriented Programe
1. encapsulation(캡슐화)
2. inheritance(상속)
3. polymorphism(다형성)



```python
class MyClass:
  cl_mem = 100 # 클래스 멤버
  cl_list = [1,2,3,4]
  a = 'hello'
```

```python
print(MyClass.cl_mem)
MyClass.cl_mem = 200
print(MyClass.cl_mem)
print(MyClass.cl_list)
print(MyClass.a)
```

```python
m1 = MyClass()
m1.a
```

## 생성자함수(메서드)
인스턴스 객체를 생성할 때 자동으로 호출이 가능
변수명 self 키워드가 고정(왼쪽 파라메터)

```python
class MyClass2:
  def __init__(self, a, b):
    print("MyClass2...")
    print(self)
```

```python
m1 = MyClass2(1, 2)
```

```python
class MyClass3:
  def __init__(self, var1):
    print("init...")
    self.in_mem = 0
    self.in_list = [0]
    self.a = var1

  def set(self, var1, var2):
    print("set....")
    self.in_mem = var1
    self.a = var2

  def get(self):
    print("get...")
    return self.in_mem, self.a

  def __del__(self):
    print("del...")
```

```python
m1 = MyClass3(100)
m1.set(30, 40)
print(m1.in_mem, m1.a)
a, b= m1.get()
print(a, b)

del m1
```
