```python
import numpy as np
```

```python
data = [[2, 0, 81],
        [4, 4, 93],
        [6, 2, 91],
        [8, 3, 97]]
```

```python
# 리스트 컴프레이션
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]
print(x1)
print(x2)
print(y)
```

```python
X1 = np.array(x1)
X2 = np.array(x2)
Y = np.array(y)
print(type(X1))
```

```python
a1 = a2 = b = 0
lr = 0.02 # 학습률
epochs = 2001 # 반복 횟수

#경사 하강법
for i in range(epochs):
  # 예측값: a1x1 + a2x2 + b
  y_pred = a1 * X1 + a2 * X2 + b

  # 오차: 실제값 - 예측값
  error = Y - y_pred

  # a1편미분
  a1_diff = (2/len(X1)) * sum(-X1 * (error))
  # a2편미분
  a2_diff = (2/len(X2)) * sum(-X2 * (error))
  # b의 편미분
  b_diff  = (2/len(X1)) * sum(-(error))

  # a1, a2, b업데이트
  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b  = b - lr * b_diff

  if i % 100 == 0:
    print("epochs=%.f, 기울기a1=%.04f, 기울기a2=%.04f, 절편=%.04f"%(i, a1, a2, b))
```

```python
y_pred = a1 * X1 + a2 * X2 + b

print("실제값: ", y)
print("예측값: ", y_pred)
```
